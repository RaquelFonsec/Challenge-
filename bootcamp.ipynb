```python
import pandas as pd
from google.colab import files
import io
from datetime import datetime

uploaded = files.upload()

# Ajuste o nome do arquivo removendo espaços
file_name = next(iter(uploaded))

# Carregue o DataFrame a partir do conteúdo do arquivo
df_concurso = pd.read_csv(io.BytesIO(uploaded[file_name]))

# 10. Quantidade de Pessoas do Estado do Piauí
quantidade_piaui = df_concurso[df_concurso['Estado'] == 'Piauí'].shape[0]
print("10. Quantidade de Pessoas do Estado do Piauí:")
print(quantidade_piaui)

# 11. Quantidade de Mulheres do Estado do Espírito Santo
quantidade_mulheres_es = df_concurso[(df_concurso['Estado'] == 'Espírito Santo') & (df_concurso['Sexo'] == 'Feminino')].shape[0]
print("11. Quantidade de Mulheres do Estado do Espírito Santo:")
print(quantidade_mulheres_es)

# 12. Quantidade de pessoas que nasceram antes de 1995
df_concurso['Ano de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'], errors='coerce').dt.year
pessoas_antes_1995 = df_concurso[df_concurso['Ano de Nascimento'] < 1995]
quantidade_antes_1995 = pessoas_antes_1995.shape[0]
print("12. Quantidade de pessoas que nasceram antes de 1995:")
print(quantidade_antes_1995)

# 13. Retorne os dados dos inscritos com as colunas: Nº Inscrição, Nome e Idade
dados_inscritos = df_concurso[['Número de Inscrição', 'Nome', 'Data de Nascimento']].copy()
dados_inscritos['Idade'] = (datetime.now() - pd.to_datetime(dados_inscritos['Data de Nascimento'], errors='coerce')).astype('<m8[Y]')
dados_inscritos = dados_inscritos[['Número de Inscrição', 'Nome', 'Idade']]
print("\n13. Dados dos inscritos com as colunas: Nº Inscrição, Nome e Idade:")
print(dados_inscritos)



```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 14. Preencher valores faltosos
df_concurso_filled = df_concurso.copy()

# Colunas do tipo string (object)
string_columns = df_concurso_filled.select_dtypes(include='object').columns
df_concurso_filled[string_columns] = df_concurso_filled[string_columns].fillna('Falta informação')

# Colunas numéricas
numeric_columns = df_concurso_filled.select_dtypes(exclude='object').columns
df_concurso_filled[numeric_columns] = df_concurso_filled[numeric_columns].fillna(df_concurso_filled[numeric_columns].median())

# 15. Plotar porcentagem de dados faltosos por coluna
missing_percentage = df_concurso_filled.isnull().mean() * 100
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_percentage.index, y=missing_percentage)
plt.title('Porcentagem de Dados Faltosos por Coluna')
plt.xlabel('Coluna')
plt.ylabel('Porcentagem de Dados Faltosos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
